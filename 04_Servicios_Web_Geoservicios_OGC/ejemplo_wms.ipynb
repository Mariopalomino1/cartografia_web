**Consultar un WMS**

```
from owslib.wms import WebMapService
wms = WebMapService('http://geocarto.igac.gov.co/geoservicios/quinientos_mil/wms', version='1.1.1')
print (wms.identification.type)
print ( wms.identification.title)
```

**Listar las capas publicadas en el servicio WMS**

```
list(wms.contents)
```

Detalles de una capa

```

print(  wms['Todas'].title )
print(  wms['Todas'].queryable )
print(  wms['Todas'].opaque )
print(  wms['Todas'].boundingBox )
print(  wms['Todas'].boundingBoxWGS84)
print(  wms['Todas'].crsOptions)
print(  wms['Todas'].styles)
```

Obtener imagen desde el servicio WMS

```
img = wms.getmap(   layers=['Todas'],
                    styles=[''],
                     srs='EPSG:4326',
                     bbox=(-74.5684,	4.4181,	-73.9964,	4.852),
                     size=(640, 480),
                     format='image/jpeg',
                     transparent=True
                     )
out = open('jpl_mosaic_visb.jpg', 'wb')
out.write(img.read())
out.close()
```
